```python
import matplotlib.pyplot as plt
import numpy as np
from von_mises_sampling import VonMisesDistAnalyzer
```

```python
samples = VonMisesDistAnalyzer.generate_von_mises_samples(mu=np.pi, kappa=1.5)
```

```python
# Display the histogram of the samples
VonMisesDistAnalyzer.display_histogram(samples)

# Construct figure and axis to plot on
VonMisesDistAnalyzer.display_polar_histogaram(samples)

```

```python
wind_directions_dec = \
    np.array([180, 300, 135, 270, 180, 180, 180, 150, 150, 180,
              150, 150, 150, 150, 180, 180, 150, 135, 150, 180,
              315, 120, 150, 150, 180, 180, 180, 150, 150, 210,
              270])
wind_directions_dec_rad = ((wind_directions_dec + 180) % 360 - 180) * np.pi / 180

VonMisesDistAnalyzer.display_polar_histogaram(wind_directions_dec_rad, N=12)
```

```python
wind_directions_jun = \
    np.array([0, 0, 0, 0, 180, 120, 300, 330, 315, 180,
              270, 300, 180, 150, 0, 210, 225, 240, 150, 60,
              270, 135, 0, 210, 180, 150, 240, 270, 150, 270])
wind_directions_jun_rad = ((wind_directions_jun + 180) % 360 - 180) * np.pi / 180

VonMisesDistAnalyzer.display_polar_histogaram(wind_directions_jun_rad, N=12)
```

```python
# Calculate mean and concentration of December wind directions
mu_dec, kappa_dec = VonMisesDistAnalyzer.calculate_mean_concentration(wind_directions_dec_rad)
print(f"mean Dec: ", mu_dec)
print(f"concentration Dec: ", kappa_dec)

# Calculate mean and concentration of December wind directions
mu_jun, kappa_jun = VonMisesDistAnalyzer.calculate_mean_concentration(wind_directions_jun_rad)
print(f"mean Jun: ", mu_jun)
print(f"concentration Jun: ", kappa_jun)
```
